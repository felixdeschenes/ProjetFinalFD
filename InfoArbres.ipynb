```python
arb = pd.read_csv("tousarbres2.csv", low_memory=False)
```

Là où «arb» désigne la lecture du fichier "tousarbres2.csv", qui comprend l'inventaire des arbres publics de Montréal. Puisque Pandas éprouvait des problèmes à lire un fichier aussi volumineux ("tousarbres2.csv" pèse près de 90Mo), j'ai ajouté la fonction "low_memory=False".

```python
arb
```

```python
arb.Remarquable.value_counts()
```

Avec la fonction "value_counts", je vérifie rapidement combien d'arbres contiennent la valeur 'O' dans la colonne "Remarquable".

```python
rema = arb.Remarquable == "O"
```

Je crée la variable "rema" pour identifier les arbres remarquables.

```python
arb[rema].Nom_arrond.value_counts()
```

À l'aide de la variable "rema", j'applique la fonction "value_counts()" uniquement aux arbres remarquables. Je peux ensuite décliner leur nombre pour chacun des arrondissements répertoriés. Cette opération a été utile pour cibler des arrondissements qui ont en réalité un nombre important de grands et de vieux arbres (comme l'arrondissement Mercier-Hochelaga-Maisonneuve), mais dont les arbres identifiés comme «remarquables» dans l'inventaire de la Ville sont très peu nombreux. Ces statistiques mises en exergue avec l'aide de Pandas offrent alors de bonnes pistes de questions à poser aux arrondissements ciblés pour comprendre la logique derrière leur évaluation des spécimens exceptionnels.
